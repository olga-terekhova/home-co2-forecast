```python
# set up paths

drive_data_dir = '/MyDrive/Projects/home-co2-forecast/data/'

drive_model_dir = '/MyDrive/Projects/home-co2-forecast/model/'
mount_dir = '/content/drive'
data_dir = mount_dir + drive_data_dir

model_dir = mount_dir+drive_model_dir
```

```python
# mount data source

from google.colab import drive
drive.mount(mount_dir)
```

```python
# import libraries

import pandas as pd
import pickle
```

```python
# read model from pickle

with open(model_dir+'model-regressor.pkl', 'rb') as f:
    loaded_model_dict = pickle.load(f)
```

```python
loaded_model_dict
```

```python
feature_list = loaded_model_dict['feature_names']
model = loaded_model_dict['model']
```

```python
# read data

X = pd.read_pickle(data_dir + 'serve_X.pkl')
```

```python
# limit the data with the feature list

X_pred = X[feature_list].head(1)
X_pred
```

```python
# predict value

y_pred = model.predict(X_pred)
y_pred
```
