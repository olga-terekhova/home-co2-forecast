```python
# import libraries

import os, mlflow, shutil
import pandas as pd
from mlflow.tracking import MlflowClient
```

```python
# set up paths

drive_data_dir = '/MyDrive/Projects/home-co2-forecast/data/'
drive_mlflow_dir = '/MyDrive/Projects/home-co2-forecast/mlflow/'
drive_model_dir = '/MyDrive/Projects/home-co2-forecast/model/'
drive_test_results_dir = '/MyDrive/Projects/home-co2-forecast/test-results/'
mount_dir = '/content/drive'
data_dir = mount_dir + drive_data_dir
mlflow_dir = mount_dir + drive_mlflow_dir
model_dir = mount_dir+drive_model_dir

DRIVE_DB = mlflow_dir + "mlflow.db"
LOCAL_DB = "/content/mlflow.db"

test_results_dir = mount_dir+drive_test_results_dir

# BACKEND_DB = mlflow_dir+"mlflow.db"
ARTIFACT_ROOT = mlflow_dir+"artifacts/"
```

```python
# mount data source

from google.colab import drive
drive.mount(mount_dir)
```

```python
# Copy the db from Drive to local disk if it already exists there.
# If it does not exist yet (first ever run), skip the copy and let
# mlflow create a fresh local db file on first write.
if os.path.exists(DRIVE_DB):
    shutil.copy2(DRIVE_DB, LOCAL_DB)
    print("Copied existing db from Drive to local disk.")
else:
    print("No db found on Drive yet. A new local db will be created.")
```

```python
# create artifact directories if needed

os.makedirs(ARTIFACT_ROOT, exist_ok=True)
```

```python
# set mlflow tracking
mlflow.set_tracking_uri("sqlite:///"+LOCAL_DB)
mlflow.get_tracking_uri()
```

```python
ex = mlflow.search_experiments()
ex
```

```python
experiments_list = [
#     'model_search_all_folds_10',
#  'model_search_all_folds_1_3',
#  'model_search_all_folds_4_6',
#  'model_search_all_folds_7_8',
#  'model_search_all_folds_9',
    'model_search_hgbr_2'
    ]
```

```python
all_experiments_runs_df = []
for ex in experiments_list:
  print(ex)
  experiment = mlflow.get_experiment_by_name(ex)
  print(experiment)
  print(experiment.experiment_id)
  experiment_runs_df = mlflow.search_runs(experiment_ids=[experiment.experiment_id])
  all_experiments_runs_df.append(experiment_runs_df)

runs_df = pd.concat(all_experiments_runs_df)
runs_df
```

```python
# save runs_df to csv in data_dir
runs_df.to_csv(test_results_dir+'runs_df.csv', index=False)
```

```python
# get 10 best rows with minimal RMSE

best_10_rmse = runs_df.nsmallest(20, 'metrics.rmse_mean')

# sort by rmse_mean
best_10_rmse = best_10_rmse.sort_values(by=['metrics.rmse_mean'])

best_10_rmse
```
